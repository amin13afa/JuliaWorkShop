# Quantum Computing with ITensors and PastaQ

```julia
using ITensors
```

<img src="img/IT2_1.png" width=400 >

```julia
d = 10;
i,j,k,l,m,n = Index.((d,d,d,d,d,d));
```

```julia
A = randomITensor(i,j,m);
B = randomITensor(j,k,n);
C = randomITensor(k,l,m);
D = randomITensor(l,i,n);
```

```julia
f(A,B,C,D) = (A*B*C*D)[]
```

```julia
using Zygote
```

```julia
∂f_∂A, ∂f_∂B, ∂f_∂C, ∂f_∂D = gradient(f,A,B,C,D);
```

```julia
typeof(∂f_∂A)
```

```julia
@time @show isapprox(∂f_∂A, B*C*D);
```

```julia
@time @show isapprox(∂f_∂B, A*C*D);
```
