# Introduction the model 

Butterfly Attractor 
$$
\frac{dx}{dt} = a(y - x)
$$
$$
\frac{dy}{dt} = x(b-z) - y
$$
$$
\frac{dz}{dt} = xy - cz
$$

We choose our vars as follows :
$$
u = \begin{bmatrix}
x \\ y \\ z
\end{bmatrix}
$$

and also 

$$
p = \begin{bmatrix}
a \\ b \\ c
\end{bmatrix}
$$

# Code the model

```julia
# Import library
using DifferentialEquations 
using Plots
```

```julia
# Define problem
function butterfly(du,u,p,t)
    a,b,c = p
    du[1] = a*(u[2] - u[1])
    du[2] = u[1]*(b - u[3]) - u[2]
    du[3] = u[1]*u[2] - c*u[3]
end;
```

```julia
# Define problem variables
u₀ = [1.0,0.0,0.0];
p = [10,28,8/3];
tspan = (0.0,100.0);
```

```julia
prob = ODEProblem(butterfly,u₀,tspan,p);
```

```julia
sol = solve(prob);
```

```julia
sol.t[13], sol[13];
```

```julia
# sol[j,i] is the value of the jth variable at time i:
sol[1,13];
```

```julia
using Plots
# define the Lorenz attractor
@kwdef mutable struct Lorenz
    dt::Float64 = 0.02
    σ::Float64 = 10
    ρ::Float64 = 28
    β::Float64 = 8/3
    x::Float64 = 2
    y::Float64 = 1
    z::Float64 = 1
end

function step!(l::Lorenz)
    dx = l.σ * (l.y - l.x);         l.x += l.dt * dx
    dy = l.x * (l.ρ - l.z) - l.y;   l.y += l.dt * dy
    dz = l.x * l.y - l.β * l.z;     l.z += l.dt * dz
end

attractor = Lorenz()

# initialize a 3D plot with 1 empty series
plt = plot3d(
    1,
    xlim = (-30, 30),
    ylim = (-30, 30),
    zlim = (0, 60),
    title = "Lorenz Attractor",
    marker = 2,
)

# build an animated gif by pushing new points to the plot, saving every 10th frame
@gif for i=1:1500
    step!(attractor)
    push!(plt, attractor.x, attractor.y, attractor.z)
end every 10
```
