Pkg.add("Plots");

```julia
using Plots
```

```julia
x = collect(-10:0.1:10);
y1 = x.^2;
y2 = x.^3;
```

```julia
plot(x,y1)
```

```julia
plot!(x,y2)
```

```julia
plot(x,y1,label="y₁=x²",title="Basic Line Plot", linewidth=3,linestyle=:dash,color=:red)
plot!(x,y2,label="y₂=x³",lw=2,ls=:dashdot,color=:blue)
xlabel!("X")
ylabel!("f(x)")
xlims!((-10,10))
ylims!(-1e3,1e3)
xticks!((-10:5:10),[string(z) for z in -10:5:10])
annotate!(0,0,"⚫")
annotate!(0,100,"Intersection")
```

```julia
theme(:dark)
plot(x->log2(x),c=:magenta, label="f(x)=log₂(x)")
xlabel!("X")
ylabel!("log₂")
ylims!(-12,8)
yticks!((-12:5:8),[string(z) for z in -12:5:8])
```

# 3D plots

```julia
x = -2:0.1:2
y = -2:0.1:2
f(x,y) = exp((-x^2-y^2))
```

```julia
color_gradient=cgrad([:violet,:blue])
surface(x,y,f,c=color_gradient)
xlabel!("X")
ylabel!("Y")
title!("3D surface plot")
```

```julia
cm = cgrad([:orange,:red])
surface(x,y,f,c=cm)
```

```julia
contour(x,y,f,levels=5,fill=true,cbar=true,c=:turbo,clabels=true)#,clabels=true) #c=:turbo # c=:inferno
annotate!(0,0,"⚫")
```

```julia
t = 0:π/100:8π
x = cos.(t)
y = sin.(t)
plot3d(x,y,t,c=:red,legend=false,lw=2)
xlabel!("X")
ylabel!("Y")
title!("3D plot")
```
